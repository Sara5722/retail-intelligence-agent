Import and load all files

```python
import pandas as pd

orders = pd.read_csv('../data/orders.csv')
order_products = pd.read_csv('../data/order_products__prior.csv')
products = pd.read_csv('../data/products.csv')
```

Explore orders

```python
# Orders table
print("ORDERS")
print(orders.shape)
print(orders.head())
print(orders.dtypes)
print(orders.isnull().sum())
```

```python
# Order products table
print("ORDER PRODUCTS")
print(order_products.shape)
print(order_products.head())
print(order_products.isnull().sum())
```

```python
# Products table
print("PRODUCTS")
print(products.shape)
print(products.head())
print(products.isnull().sum())
```

join

```python
# Join order_products with product names
order_products_named = order_products.merge(products, on='product_id', how='left')

# Join with order info (customer, day of week, etc.)
full_df = order_products_named.merge(orders, on='order_id', how='left')

print(full_df.shape)
print(full_df.head())
```
